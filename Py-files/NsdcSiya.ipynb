```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/NSDC - World Energy Consumption.csv')
df
```

```python
df.info()
```

Data Cleaning

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df= df.drop_duplicates()
```

```python
df= df.replace('-', np.nan)
df
```

```python
df=df.drop(df.index[0])
```

```python
df.info()
```

```python
df.iloc[:, 1:] = df.iloc[:, 1:].bfill(axis=1)
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(float)
```

```python
df = df.sort_values(by='Country').reset_index(drop=True)
df
```

```python
df['Mean_Energy'] = df.iloc[:, 1:].mean(axis=1).astype(float).round(3)
df
```

```python
import pycountry_convert as pc
def get_continent(country_name):
    try:
        country_code = pc.country_name_to_country_alpha2(country_name, cn_name_format="default")
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_map = {
            "AF": "Africa",
            "AS": "Asia",
            "EU": "Europe",
            "NA": "North America",
            "SA": "South America",
            "OC": "Oceania"
        }
        return continent_map[continent_code]
    except:
        return "Other"
df["Continent"] = df["Country"].apply(get_continent)
```

```python
df = df.set_index('Country')
```

```python
df.index = df.index.str.strip()
rows_to_drop = ["OECD", "Non-OECD", "European Union #", "of which: OECD","USSR","Total World"]
df = df.drop(rows_to_drop, errors='ignore')
```

```python
df["1965-1975"] = df.loc[:, "1965":"1975"].mean(axis=1).astype(float).round(3)
df["1976-1985"] = df.loc[:, "1976":"1985"].mean(axis=1).astype(float).round(3)
df["1986-1995"] = df.loc[:, "1986":"1995"].mean(axis=1).astype(float).round(3)
df["1996-2005"] = df.loc[:, "1996":"2005"].mean(axis=1).astype(float).round(3)
df["2006-2015"] = df.loc[:, "2006":"2015"].mean(axis=1).astype(float).round(3)
df["2016-2023"] = df.loc[:, "2016":"2023"].mean(axis=1).astype(float).round(3)
df = df.drop(columns=df.loc[:, "1965":"2023"].columns)
df
```

```python
plt.figure(figsize=(12, 6))

df_reset = df.reset_index()
df_reset["Growth"] = df_reset["2016-2023"] - df_reset["1965-1975"]
top_growth = df_reset.sort_values("Growth", ascending=False).head(10)

plt.bar(top_growth["Country"], top_growth["Growth"])
plt.xlabel("Country")
plt.ylabel("Growth (Last – First Decade)")
plt.title("Top 10 Growth Countries")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y")
plt.show()
```

```python
plt.figure(figsize=(12, 6))

continents = df['Continent'].unique()
decades = ["1965-1975","1976-1985","1986-1995","1996-2005","2006-2015","2016-2023"]

for cont in continents:
    data = df[df['Continent'] == cont][decades].mean()
    plt.plot(decades, data, marker='o', label=cont)

plt.xlabel("Decade")
plt.ylabel("Avg Energy Consumption")
plt.title("Continent Energy Trends Over Decades")
plt.xticks(rotation=45)
plt.legend(fontsize=9, bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```
