```python
df_median=df['bath'].median()
df['society']=df['society'].fillna(method='ffill').fillna(method='bfill')
df=df.fillna({'balcony':0,'bath':df_median})
df=df.fillna({'size':'unknown', 'location':'unknown'})
print(df.head(10))
print(df.isnull().sum())
```

```python
print("Sorting using price")
df_sort = df.sort_values(by= 'price', ascending= True)
df_sort
```

```python
print("Sorting using area")
df_new= df.sort_values(by='balcony', ascending= True)
df_new
```
