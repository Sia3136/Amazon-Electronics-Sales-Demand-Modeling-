Data Visualization

```python
plt.figure(figsize=(8,6))
plt.scatter(df['bmi'], df['blood_glucose_level'], c=df['diabetes'], cmap='coolwarm', alpha=0.3)
plt.colorbar(label='Diabetes')
plt.xlabel('BMI')
plt.ylabel('Blood Glucose Level')
plt.title('BMI vs Blood Glucose Level')
plt.show()
```

It shows that people with higher blood glucose have higher diabetes range

```python
plt.figure(figsize=(8,6))
plt.scatter(df['bmi'], df['hbA1c_level'], c=df['diabetes'], cmap='coolwarm', alpha=0.3)
plt.colorbar(label='Diabetes')
plt.xlabel('BMI')
plt.ylabel('HbA1c Level')
plt.title('BMI vs HbA1c Level')
plt.show()
```

It shows that people with higher HbA1c levels have more diabtes

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[['age','bmi','hbA1c_level','blood_glucose_level','diabetes']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

This heatmap shows relation among different entities

```python
df['age_group'] = pd.cut(df['age'], bins=[0,20,40,60,80,100],
                         labels=['0-20','21-40','41-60','61-80','81+'])
age_diabetes = df.groupby('age_group')['diabetes'].mean().reset_index()
plt.figure(figsize=(7,5))
sns.barplot(x='age_group', y='diabetes', data=age_diabetes, palette='Blues')
plt.title('Diabetes Prevalence by Age Group')
plt.ylabel('Proportion with Diabetes')
plt.xlabel('Age Group')
plt.show()
```

Here we grouped the columns , so to find which age are prone to diabtes

```python
df['bmi_category'] = pd.cut(df['bmi'], bins=[0,18.5,25,30,40,100],
                            labels=['Underweight','Normal','Overweight','Obese','Severely Obese'])

bmi_diabetes = df.groupby('bmi_category')['diabetes'].mean().reset_index()

plt.figure(figsize=(7,5))
sns.barplot(x='bmi_category', y='diabetes', data=bmi_diabetes, palette='Reds')
plt.title('Diabetes Prevalence by BMI Category')
plt.ylabel('Proportion with Diabetes')
plt.xlabel('BMI Category')
plt.show()
```

Here we grouped the columns of weights to categories , and we can see severly obese people have higer chances of being diabetic

Ml models

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
tree= DecisionTreeClassifier( random_state=41)
```

```python
from sklearn.preprocessing import LabelEncoder
cols_to_encode = ['gender', 'smoking_history', 'location']
le = LabelEncoder()
for col in cols_to_encode:
    df[col + '_encoded'] = le.fit_transform(df[col])
x = df[['age', 'gender_encoded', 'hypertension', 'heart_disease',
        'smoking_history_encoded', 'bmi', 'hbA1c_level', 'blood_glucose_level',
        'race:AfricanAmerican', 'race:Asian', 'race:Caucasian',
        'race:Hispanic', 'race:Other', 'location_encoded']]
y = df['diabetes']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 41)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(x_train, y_train)
y_pred_lr = log_reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Precision:", precision_score(y_test, y_pred_lr))
print("Recall:", recall_score(y_test, y_pred_lr))
print("F1 Score:", f1_score(y_test, y_pred_lr))
```
