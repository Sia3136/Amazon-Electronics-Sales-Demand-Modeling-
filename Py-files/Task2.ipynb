```python
import numpy as np
import random
```

```python
np.random.seed(5)
r = np.random.randint(1,21,(6,6))
print(r)
```

```python
mean = r.mean(axis=0)
print(mean)
print()
res = r - mean
print(res)
```

```python
p = res[abs(res) >= 1.5]
print(p)
```

```python
print(np.sum(p))
print(np.mean(p))
print(np.std(p))
print(np.min(p))
print(np.max(p))
```

```python
np.random.shuffle(r)
print(r[:3])
```
